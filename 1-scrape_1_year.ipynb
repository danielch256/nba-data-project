## 1. Scrape player stats of 1 year from basketball-reference.com

```python
# imports
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
```

```python
# open url and save as soup object
url = 'https://www.basketball-reference.com/leagues/NBA_2016_advanced.html'
url_data = urlopen(url)
url_html = url_data.read()
url_data.close()

my_soup = soup(url_html, 'html.parser')
```

```python
# find all html tables
soup_table = my_soup.findAll('table') #prints out an array of all tables, which in this case contains only 1
table = soup_table[0] # the table we want is at index 0 in the array
table_headers = table.findAll('th', {}) # find all headers in the table
```

```python
header_titles = [] # square brackets means its an array. why do i make it an array? idk
for header in table_headers:
  header_titles.append(header.text)
header_titles_trimmed = header_titles[1:29]
header_titles_trimmed # these are the texts in the header
```

```python
all_rows = my_soup.findAll('tr', {}) # this returns an array, first index is the headers
rows = all_rows[1:] #get rid of headers
first_row = rows[0]
first_row_data = first_row.findAll('td', {})
```

```python
row_text_all = [] 
# get the text from all the rows
for row in rows:
  row_text=[]
  row_data = row.findAll('td', {})
  for td in row_data:
    row_text.append(td.text)
  row_text_all.append(row_text)
row_text_all
```

```python
filename = '2016.csv' #save the header text and row text in a csv for the year 2016
f = open(filename, 'w')
header_string = ''
for title in header_titles_trimmed: #extract header content as a string and format for csv
  header_string += title + ','
header_string = header_string[:-1]
header_string += '\n'
f.write(header_string)

for row in row_text_all: #same for body
  row_string = ''
  for td in row:
    row_string += td + ','
  row_string = row_string[:-1]
  row_string += '\n'
  f.write(row_string)
  
```
